# Linear Algebra


### Vector --->   it is collection of list or array of numbers. It represents both magnitude and direction. it can be either 1row or 1column. to perform operations b/w two vectors, the shape of two vectors should be same ..

### Scaler ---> It is nothing but a Value . it represent only magnitude.

#### Vector Operations

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

##### plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,-2,3,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Addition of vectors
vector1 = np.array([4,5])
vector2 = np.array([-2,3])
vector_sum = vector1 + vector2
print(vector_sum)
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-2,3,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,vector_sum[0],vector_sum[1],scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-8,8)  
plt.ylim(-8,8)
plt.show()
```

```python
#subtraction of vectors
vector1 = np.array([2,3])
vector2 = np.array([3,-2])
vector_diff = vector1 - vector2
print(vector_diff)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,vector_diff[0],vector_diff[1],scale_units='xy',angles='xy',scale=1,color='r')
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.show()
```

```python
#Multiplication of vector by a scalar
vector = np.array([2,3])
scalar = 3
scaled_vector = scalar * vector
print(scaled_vector)
```

```python
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,6,9,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
#Dot product of vectors
vector1 = np.array([1,2])
vector2 = np.array([3,4])
dot_product = np.dot(vector1, vector2)  # --> (1*3) + (2*4) = 3 + 8 = 11
print(dot_product)
```

```python
#Cross product of vectors
vector1 = np.array([1,2,3])
vector2 = np.array([4,5,6])
cross_product = np.cross(vector1, vector2)  # --> [ (2*6 - 3*5), (3*4 - 1*6), (1*5 - 2*4) ] = [-3, 6, -3]
print(cross_product)
```

```python
#projection of vector1 onto vector2
vector1 = np.array([2,5])
vector2 = np.array([8,-6])
projection = (np.dot(vector1, vector2) / np.dot(vector2, vector2)) * vector2  # --> (2*8 + 5*(-6)) / (8*8 + (-6)*(-6)) * [8,-6] = (16-30) / (64+36) * [8,-6] = -14/100 * [8,-6] = [-1.12, 0.84]
print(projection)
```

```python
#projection of vector1 onto vector2
vector1 = np.array([23,45,62])
vector2 = np.array([45,82,67])
projection = (np.dot(vector1, vector2) / np.dot(vector2, vector2)) * vector2  #--> (23*45 + 45*82 + 62*67) / (45*45 + 82*82 + 67*67) * [45,82,67] = (1035 + 3690 + 4154) / (2025 + 6724 + 4489) * [45,82,67] = 8889 / 13238 * [45,82,67] = [30.21, 55.03, 44.91]
print(projection)
```
