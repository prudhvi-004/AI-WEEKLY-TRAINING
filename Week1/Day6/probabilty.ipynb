# Probability

- it is a branch of mathematics that deals with calculating the likelihood of a given event to occur.



```python
import numpy as np
```

#### Conditional Probability

- Probability of an event given another event has occurred
- Depends on relationship between events

```python
# # Example: students passing math given they studied
studied = np.array([1,1,0,1,0])
passed  = np.array([1,0,0,1,0])

P_B = np.mean(studied)
P_A_and_B = np.mean((studied == 1) & (passed == 1))

P_A_given_B = P_A_and_B / P_B
print("P(A|B):", P_A_given_B)
```

#### Bayes Theorm

- Used to update probability based on new evidence
- Widely used in ML classification problems

```python
# Example values
P_A = 0.3
P_B_given_A = 0.8
P_B = 0.5

P_A_given_B = (P_B_given_A * P_A) / P_B
print("Bayes Result:", P_A_given_B)
```

#### Random Variables
- Variable whose value depends on random outcomes
- Can be discrete or continuous

```python
dice = np.random.randint(1, 7, size=10)
print("Random values:", dice)
```

#### Probability Distributions
- Describe how probabilities are distributed over values
- Common types: Normal, Uniform, Binomial

```python
normal = np.random.normal(0, 1, 5)
print("Normal Distribution:", normal)
```

#### Independence of Events
- Two events are independent if one does not affect the other

```python
P_A = 0.5
P_B = 0.4

print("P(A ∩ B):", P_A * P_B)
```

#### Joint & Marginal Probability
- Joint probability measures combined events
- Marginal probability is individual probability

```python
joint = np.array([[0.1, 0.2],
                  [0.3, 0.4]])

print("Marginal A:", np.sum(joint, axis=1))
print("Marginal B:", np.sum(joint, axis=0))
```

#### Normal Distribution (Gaussian)
- A continuous distribution that is symmetric and bell-shaped
- Defined by mean (μ) and standard deviation (σ)

```python
# Generate normal distribution data
data = np.random.normal(loc=0, scale=1, size=10)

print("Normal Distribution:", data)
```

#### Poisson Distribution
- A discrete distribution used for counting events in a fixed interval
- Defined by λ (lambda) = average rate of occurrence

```python
# Generate poisson distribution data
data = np.random.poisson(lam=3, size=10)

print("Poisson Distribution:", data)
```
