# Statistics

- It is the science concerned with developing and studying methods for collecting, analyzing, interpreting and presenting data.
- It is a tool that helps us to extract information & knowledge from data.

### Types Of Statistics

#### 1. Descriptive Statistics 

- they are used to describe the basic features of the data ina study. they provide simple summarises about the sample and the measures.
- It describes the Data. 
- Collecting ,summarizing & presenting data.
- Data Analysis.
- Measure of central tendencies(Mean,Median,Mode)
- Measure of Variability(Range, SD, Variance)

#### 2. Inferential Statistics

- It takes data from a sample and makes inferences and predictions about the larger population from which the sample was drawn.
- Get inference about the data.
- making predictions , generalization or decisions based on data.
- Correlation

#### Statistical Measures 

#####  Measure of Central Tendency 

- It represents the central or typical value of a dataset
- It helps to summarize large data into a single value

1. Mean
- It is the average of all values in a dataset
- It is affected by outliers (extreme values)
2. Median
- It is the middle value when data is arranged in order
- It is not affected by outliers
3. Mode
- It is the most frequently occurring value
- It is useful for categorical data

##### Measure of Variability

1. Range
- Difference between maximum and minimum values
- Simple but affected by outliers
2. Variance
- Measures how far values are from the mean (spread)
- Uses squared differences, so always positive
3. Standard Deviation
- Square root of variance
- Shows average deviation from the mean

```python
import numpy as np
```

```python
# to calculate Mean

data = np.array([10, 20, 30, 40, 50])

mean = np.mean(data)
print("Mean:", mean)
```

```python
#to calculate Median
median = np.median(data)
print("Median:", median)
```

```python
#to calculate Mode
data = np.array([10, 20, 20, 30, 40])

values, counts = np.unique(data, return_counts=True)
mode = values[np.argmax(counts)]

print("Mode:", mode)
```

```python
# to calculate Standard Deviation
std_dev = np.std(data)
print("Standard Deviation:", std_dev)
```

```python
#to calculate Range
data = np.array([10, 20, 30, 40, 50])
data_range = np.ptp(data)
print("Range:", data_range)
```

```python
#to calculate variance
variance = np.var(data)
print("Variance:", variance)
```
