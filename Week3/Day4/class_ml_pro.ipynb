```python
# separate the data and Label
X = df.drop('quality',axis=1)
```

```python
X
```

```python
Y = df['quality'].apply(lambda y_value: 1 if y_value>=6 else 0)
```

```python
Y
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=40)
```

```python
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score
```

```python
def model_eval(true,pred):
    acc = accuracy_score(true,pred)
    pre = precision_score(true,pred)
    reca = recall_score(true,pred)
    f1 = f1_score(true,pred)

    return acc,pre,reca,f1
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
models = [KNeighborsClassifier(),DecisionTreeClassifier(),SVC(),RandomForestClassifier(),XGBClassifier()]
```

```python
from sklearn.model_selection import cross_val_score
```

```python
for model in models:
    model.fit(X_train,Y_train)

    X_train_pred = model.predict(X_train)
    X_test_pred = model.predict(X_test)

    model_train_acc,model_train_pre,model_train_reca,model_train_f1 = model_eval(X_train_pred,Y_train)

    model_test_acc,model_test_pre,model_test_reca,model_test_f1 = model_eval(Y_test,X_test_pred)


    print("Model:",model)
    print()

    print(f"Model Train MSE:{model_train_acc}")
    print(f"Model Train MAE:{model_train_pre}")
    print(f"Model Train R2:{model_train_reca}")
    print(f"Model Train R2:{model_train_f1}")

    print("---------------------------------------------------------------------------")

    print(f"Model Test MSE:{model_train_acc}")
    print(f"Model Test MAE:{model_train_pre}")
    print(f"Model Test R2:{model_train_reca}")
    print(f"Model Test R2:{model_train_f1}")

    print("----------------------------------------------------------------------------")

    cv_train_score = cross_val_score(model,X_train,Y_train,cv=5)
    cv_test_score = cross_val_score(model,X_test,Y_test,cv=5)
    print("Cross Validation:",cv_train_score)
    print("Average Cross Validation Score:",np.mean(cv_train_score))
    print("Cross Validation:",cv_test_score)
    print("Average Cross Validation Score:",np.mean(cv_test_score))

    print("-----------------------------------------------------------------------------")
    print("------------------------------------------------------------------------------")
    print()
```

```python
models = [
    {
        "name": "RandomForestClassifier",
        "model": RandomForestClassifier(),
        "params": {
            "n_estimators": [100, 200, 300],
            "max_depth": [None, 10, 20],
            "min_samples_split": [2, 5],
            "criterion": ["gini", "entropy"]
        }
    },
    {
        "name": "XGBClassifier",
        "model": XGBClassifier(),
        "params": {
            "n_estimators": [100, 200],
            "learning_rate": [0.01, 0.1, 0.2],
            "max_depth": [3, 5, 7],
            "subsample": [0.8, 1],
            "colsample_bytree": [0.8, 1]
        }
    }
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

results = []

for item in models:
    print(f"Running: {item['name']}")

    search = RandomizedSearchCV(
        estimator=item["model"],
        param_distributions=item["params"],
        n_iter=10,
        cv=5,
        scoring="r2",   # regression metric
        n_jobs=-1,
        random_state=42
    )

    search.fit(X_train, Y_train)

    results.append({
        "model": item["name"],
        "best_score": search.best_score_,
        "best_params": search.best_params_
    })
```

```python
results
```

```python
best_model = search.best_estimator_
```

```python
best_model
```

```python
input_df = X.iloc[[-1]]

prediction = best_model.predict(input_df)

print(prediction)

if prediction[0] == 1:
    print("Good Quality")
else:
    print("Bad Quality")
```
