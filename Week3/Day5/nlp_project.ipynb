##Feature Engineering

```python
# To get and add the length of question in a DataFrame


df['q1_len'] = df['question1'].str.len()
df['q2_len'] = df['question2'].str.len()
```

```python
df.head()
```

```python
# To Know and Add the no of words in a Question

df['q1_num_words'] = df['question1'].apply(lambda row: len(row.split(" ")))
df['q2_num_words'] = df['question2'].apply(lambda row: len(row.split(" ")))
```

```python
df.head()
```

```python
# To Add the No of common words from question 1&2

def common_words(row):
  w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
  w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))

  return len(w1 & w2)
```

```python
df['word_common'] = df.apply(common_words, axis=1)
```

```python
df.head()
```

```python
# Total Words

def total_words(row):
  w1 = set(map(lambda word: word.lower().strip(), row['question1'].split(" ")))
  w2 = set(map(lambda word: word.lower().strip(), row['question2'].split(" ")))

  return (len(w1) + len(w2))
```

```python
df['word_total'] = df.apply(total_words, axis=1)
```

```python
df.head()
```

```python

df['word_share'] = round(df['word_common']/df['word_total'],2)
df.head()
```

```python
ques_df = df[['question1','question2']]
ques_df.head()
```

```python
final_df = df.drop(columns=['id','qid1','qid2','question1','question2'])
print(final_df.shape)
final_df.head()
```

## Using Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
questions = list(ques_df['question1']) + list(ques_df['question2'])
```

```python
cv = CountVectorizer(max_features=3000)
```

```python
q1_arr , q2_arr = np.vsplit(cv.fit_transform(questions).toarray(),2)
```

```python
temp_df1 = pd.DataFrame(q1_arr, index= ques_df.index)
temp_df2 = pd.DataFrame(q2_arr, index= ques_df.index)
temp_df = pd.concat([temp_df1, temp_df2], axis=1)
temp_df.shape
```

```python

final_df = pd.concat([final_df, temp_df], axis=1)
print(final_df.shape)
final_df.head()
```

##Evaluation

```python
X = final_df.drop(columns=['is_duplicate'])
y = final_df['is_duplicate']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

##bcs we have numbers as the column names
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
print('Test Acc',accuracy_score(y_test,y_pred))
accuracy_score(y_test,y_pred)
```
