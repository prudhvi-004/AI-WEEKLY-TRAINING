```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

def model_eval(true,predicted):
    mse = mean_squared_error(true,predicted)
    mae = mean_absolute_error(true,predicted)
    r2 = r2_score(true,predicted)

    return mse ,mae, r2
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
models = [LinearRegression(),KNeighborsRegressor(),DecisionTreeRegressor(),RandomForestRegressor(),SVR(),XGBRegressor()]
```

```python
for model in models:
    model.fit(X_train,Y_train)

    X_train_pred = model.predict(X_train)
    X_test_pred = model.predict(X_test)

    model_train_mse,model_train_mae,model_train_r2 = model_eval(X_train_pred,Y_train)

    model_test_mse,model_test_mae,model_test_r2 = model_eval(Y_test,X_test_pred)


    print("Model:",model)
    print()

    print(f"Model Train MSE:{model_train_mse}")
    print(f"Model Train MAE:{model_train_mae}")
    print(f"Model Train R2:{model_train_r2}")

    print("---------------------------------------------------------------------------")

    print(f"Model Test MSE:{model_test_mse}")
    print(f"Model Test MAE:{model_test_mae}")
    print(f"Model Test R2:{model_test_r2}")

    print("----------------------------------------------------------------------------")

    cv_train_score = cross_val_score(model,X_train,Y_train,cv=5)
    cv_test_score = cross_val_score(model,X_test,Y_test,cv=5)
    print("Cross Validation:",cv_train_score)
    print("Average Cross Validation Score:",np.mean(cv_train_score))
    print("Cross Validation:",cv_test_score)
    print("Average Cross Validation Score:",np.mean(cv_test_score))

    print("-----------------------------------------------------------------------------")
    print("------------------------------------------------------------------------------")
    print()
```

```python
models = [
    {
        "name": "RandomForest",
        "model": RandomForestRegressor(),
        "params": {
            "n_estimators": [100, 200, 300],
            "max_depth": [None, 10, 20],
            "min_samples_split": [2, 5]
        }
    },
    {
        "name": "XGBoost",
        "model": XGBRegressor(objective="reg:squarederror"),
        "params": {
            "n_estimators": [100, 200],
            "learning_rate": [0.01, 0.1, 0.2],
            "max_depth": [3, 5, 7]
        }
    }
]
```

```python
from sklearn.model_selection import RandomizedSearchCV

results = []

for item in models:
    print(f"Running: {item['name']}")

    search = RandomizedSearchCV(
        estimator=item["model"],
        param_distributions=item["params"],
        n_iter=10,
        cv=5,
        scoring="r2",   # regression metric
        n_jobs=-1,
        random_state=42
    )

    search.fit(X_train, Y_train)

    results.append({
        "model": item["name"],
        "best_score": search.best_score_,
        "best_params": search.best_params_
    })
```

```python
results
```

```python
best_model = search.best_estimator_
```
